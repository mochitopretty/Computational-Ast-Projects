<h1 style="color:hotpink; font-size:15px;"> CCD images </h1>

Through a CCD with 16 bits per pixel (in the header), referring  Charge Coupled Device, those are plane plates divided by pixels. Pixels receive light (photons), producing photoelectric effect where energy excites the system, electrons are extracted and fields directionated. 1 pixel has 3 ports with variant potential (p1 and p2 has high V, p3 has low V), those are ways in which electrons transport, the fit into potential barriers and there is amount of electrons/pixel. When electrons are read, p1 lows its potential, while in p2 and p3 increase, then we have a readout that mathematically is a matrix, and electrons convert into A/D or ADU (Analog Digital Unit).

So, brief review of the process: Photons -> electrons -> ADUs

For FITS images, we have 4 types: Bias Frames, Dark Frames, Flat Fields and Light Frames, which are described section by section. 

When we study the concepts, theoretically we can consider as an example a CCD linearity curve (the image below is an example of CCD Handbook, <div style="text-align: center;">

in which in y-axis we have the output signal in ADU, while the x-axis corresponds to detected photons. The linearity exists until ~150000 electrons (as pixel FWC) equivalent to 32767 ADU, with and output of ~26000 ADU. The linearity curve slope is the gain for the case from the CCD handbook, then where the non-linearity begins, there is limitation for the largest usable output ADU value. The same properties as FWC, output signal, the linearity, and more, we have for the context of the OMA observatory.


<h1 style="color:hotpink; font-size:15px;"> Conceptual Features </h1> Important features are described to understand the description of the different sections in the OMA observatory context. In the first place we have;
<h1 style="color:purple; font-size:13px;"> Quantum efficiency (QE): </h1> Silicon CCD capacity to absorb photons in specific wavelenghts, quantifying lost and electron recombination, depending on material absorption capacity. In short terms it is how efficient the receive electrons lecture is. QE= total electrons / initial photons. For the OMA observatory's sensor, the QE is highly efficient with high sensitivity for fast image acquisition. 
<div style="text-align: center;">
    <img src="QE_oma.png" width="500" style="border-radius:10px;">
</div>
The QE curve for the KAF16803 sensor includes lost of photons due to barrier structure, electrons recombination, surface reflection and poor absorption. At 60% occurs the maximum QE, from ~62% to 90% of QE efficiency (y-axis) information is lost, also for low wavelengths (x-axis) 70% of photons or more are reflected, while for aproximately 470 nm to 700 nm (the optical range) the sensor is much more efficient, after 700 nm or higher wavelengths occurs the mirror absorption curve.
<h1 style="color:purple; font-size:13px;"> Full Well capacity (FWC): </h1> Maximum number of electrons that 1 pixel can capture before it saturates. In the case of OMA observatory sensor KAF-16803, the FWC is about 100,000 electrons.
<h1 style="color:purple; font-size:13px;"> CCD gain (G): </h1> It refers to total number of electrons a pixel can hold per ADU (useful electrons to use in the analog digital unit), the OMA observatory FLI ProLine 16803 camera has a gain of 1.45 electrons/ADU (the common values vary between 1 to 150 e/ADU or more). This value is useful to obtain the saturation level as FWC/G, so, 100000 electrons/1.45 electrons/ADU = 68965 ADU.
<h1 style="color:purple; font-size:13px;"> Digital ADU range </h1>
The maximum out number is given by the number of bits used, (2^(#bits))-1, in this case bits per data pixel is 16, so the digital range is 0-65535 ADU . But physically the saturation point is ~68965 ADU, then the general digital saturation is 65535 ADU.
<h1 style="color:purple; font-size:13px;"> Binning: </h1> Grouping measured data into specific ranges, it combines adjacent pixels to form larger pixels and can be worked as arrays (matrix way).
<h1 style="color:purple; font-size:13px;"> Readout Noise: </h1> It is the electronics noise, due to photons beat the plate. Kind of quantified interference as number of electrons per acumulative pixels, as final device signal, e.g: the electronincs can introduce false electrons to the system.
<h1 style="color:purple; font-size:13px;"> Dark current </h1> This is the current when light is abscent, indicates how many electrons are in each pixel per exposure time. For the OMA observatory the dark current noise is obtained following a Poisson statistics.
<h1 style="color:purple; font-size:13px;"> Background noise </h1> This is the current when light is abscent, indicates how many electrons are in each pixel per exposure time.
<h1 style="color:hotpink; font-size:13px;"> Last... </h1> 
if an empirical graphic would be done too see linearity, in the case of OMA observatory's sensor the non-linearity is <1%, it is before the saturation point (at ~65535 ADU) e.g at 62000 ADU, having into account its FWC about ~100000 and the gain 1.45 electrons/ADU. <h1 style="color:hotpink; font-size:15px;"> Masterbias construction </h1>

```python
#Combination of the bias frames to create a Master Bias, and plot of Master Bias Image
#Here I used a model from the recommended website: https://docs.astropy.org/en/latest/io/fits/usage/image.html
#First each file is called
hdul1 = fits.open('image1.fit')
hdul4 = fits.open('image4.fit')
hdul6 = fits.open('image6.fit')
hdul7 = fits.open('image7.fit')
hdul10 = fits.open('image10.fit')
hdul12 = fits.open('image12.fit')
hdul15 = fits.open('image15.fit')
hdul16 = fits.open('image16.fit')
hdul18 = fits.open('image18.fit')
output = np.zeros((4096, 4096)) #it refers to the matrix with symmetric dimension 4096x4096
for i in range(41): #for the range I consider 4096/100, where 100 are bins
    j = i * 100 #this can be interpreted as the advance of each 100 rows in the image 
    k = j + 100 #this would be the consequent process for rows 
    x1 = hdul1[0].section[j:k,:] #all of the x process each 100 rows by parts, in order to construct and combine the images
    x2 = hdul4[0].section[j:k,:]
    x3 = hdul6[0].section[j:k,:]
    x4 = hdul7[0].section[j:k,:]
    x5 = hdul10[0].section[j:k,:]
    x6 = hdul12[0].section[j:k,:]
    x7 = hdul15[0].section[j:k,:]
    x8 = hdul16[0].section[j:k,:]
    x9 = hdul18[0].section[j:k,:]
    #it is NOT the same as binning, because the binning combines various pixels in one 
    #this process is just to group each 100 rows and subsequently obtain the combination
    output[j:k, :] = np.median([x1, x2, x3,x4,x5,x6,x7,x8,x9], axis=0) #median is convenient to exclude for example edges, which is noise, 
#also the median is used because data does not get affected too much if some pixel is moved up, down or to one side, instead the mean has much more sensitivity
#Cosmetic effects in general (stuck bits, hot pixels, bad pixels) can affect the mean (as sais before)
```

```python
#This file .fits can be saved to create an histogram of the final image using 100 bins
#here "output" is the master bias created above, and overwrite=True is a convenient way to simply overwrite if some change is made
fits.writeto("reduction_project/products/master_bias/master_bias.fits", output, overwrite=True)
```

<h1 style="color:hotpink; font-size:15px;"> Masterdark construction </h1>

```python
#Combination of dark frames to create a Master Dark
#Here the idea is to access directly to all the files contained in the folder we want to extract information
folder_dark = "reduction_project/dark/"
l_dark = []
bit_pixel_image = []
for file in os.listdir(folder_dark):
 #First we look into all the dark frames
    with fits.open(file) as hdul:
        dark_data = hdul[0].data
        header_dark = hdul[0].header
        exptime_dark = header_dark["EXPTIME"]
        plt.figure(figsize=(2,1.25))
        plt.imshow(dark_data, cmap="inferno",vmin=np.percentile(dark_data,5),vmax=np.percentile(dark_data,95))
        plt.colorbar()
        plt.title(f"Dark frame {file}")
        plt.show()
        plt.figure(figsize=(2,1))
        plt.hist(dark_data.flatten(),bins=30,color="palevioletred",range=(0,2000))        
        plt.title(f"Distribution for {file}")
        plt.xlabel("[ADU/sec]")
        plt.ylabel("Number of pixels")
        plt.show()
        master_dark_initial = (dark_data - master_bias)/exptime_dark 
        #in the equation we need the exposure time because CCD receives in one way photonos from background, even with the shutter closed, also theoretically the Master Dark is measured as ADU/sec, it is the total analog digital units per second 
        l_dark.append(master_dark_initial)
combination_all = np.array(l_dark)
master_dark_all_frames = np.median(combination_all, axis=0)
plt.figure(figsize=(4,3))
plt.imshow(master_dark_all_frames, cmap="inferno",vmin =np.percentile(master_dark_all_frames,5),vmax=np.percentile(master_dark_all_frames,95))
plt.title("Master Dark with 3 dark frames")
plt.colorbar()
plt.show()
print("CCD-TEMP for image2.fit:",header_dark["CCD-TEMP"],"(justification for more saturation of image2.fit)")
#When we look into the images below, we see that image2.fit (last one) is so saturated, so it has more thermal noise, 
#it is because looking into its header the CCD-TEMP keyword indicates -15 °C when normally in the others is about -26°C, 
#CCD-TEMP for image2.fit is hotter that the other two images, converting the final Master Dark not properly at all (really saturated) as seen below
#Also, that is why the histogram range for image2.fit is different from image11.fit and image8.fit
```

<h1 style="color:hotpink; font-size:13px;"> Analysis </h1>

```python
#By the justification above, just image11.fit and image8.fit are chosen to create the Master Dark
folder_dark = "reduction_project/dark/"
l_dark = []
exp_times = []
for file in os.listdir(folder_dark):
    if file != "image2.fit": 
        with fits.open(file) as hdul:
            dark_data = hdul[0].data
            header_dark = hdul[0].header
            exptime_dark = header_dark["EXPTIME"]
            exp_times.append(exptime_dark) #to obtain the median for the dark current noise 
            plt.figure(figsize=(2,1.25))
            plt.imshow(dark_data, cmap="inferno",vmin=np.percentile(dark_data,5),vmax=np.percentile(dark_data,95))
            plt.colorbar()
            plt.title(f"Dark frame {file}")
            plt.show()
            plt.figure(figsize=(2,1))
            plt.hist(dark_data.flatten(),bins=30,color="palevioletred",range=(970,1030))
            maximum_value = np.max(dark_data)        
            plt.title(f"Distribution for {file}")
            plt.xlabel("[ADU/sec]")
            plt.ylabel("Number of pixels")
            plt.show()
            print("Maximum ADU/sec for",file,":",maximum_value)
            master_dark_initial = (dark_data - master_bias)/exptime_dark
            bit_pixel_image.append([file,hdul[0].header["BITPIX"]])
            l_dark.append(master_dark_initial) #all master darks included in a list
#we have respective arrays but we need to have one combined, as following the idea (in short) for Master Bias 
combination_md = np.array(l_dark)
exp_times_mean = np.mean(np.array(exp_times),axis=0)
master_dark = np.median(combination_md, axis=0)
fits.writeto("reduction_project/products/master_dark/master_dark.fits", master_dark, overwrite=True)
plt.figure(figsize=(4,3))
plt.imshow(master_dark, cmap="inferno",vmin =np.percentile(master_dark,5),vmax=np.percentile(master_dark,95))
plt.title("Master Dark Image")
plt.colorbar()
plt.show()
```

The master dark image with 2 dark frames clearly shows thermical noise, as saturated points in a non-homogeneous background, in comparison with the master dark with 3 dark frames, this one is "oranger" (showing more saturation).

```python
#Master Dark with a proper image classification
mean_dark = np.mean(master_dark)
std_dark = np.std(master_dark) #ADU/sec
dc_noise = (std_dark**2)/exp_times_mean #is in ADU
plt.figure(figsize=(3,4))
plt.hist(master_dark.flatten(),bins=30,color="palevioletred",range=(-0.078,0.090))
plt.axvline(mean_dark, color="black", linestyle="-.", linewidth=2, label=f"$\\mu$: {round(mean_dark,2)}") 
plt.title("Distribution of Master Dark")
plt.xlabel("[ADU/sec]")
plt.ylabel("Number of pixels")
plt.legend()
plt.show()
```

<h1 style="color:hotpink; font-size:18px;"> ⋆⭒˚.⋆ Part 3: Build the Normalized MasterFlat ⋆⭒˚.⋆  </h1>

<h1 style="color:hotpink; font-size:13px;"> Data Inspection </h1> 
To create the normalized Master Flat, first we have to inspect the features of Flat Fields. Starting with the general Flat Fielding aproach, it means to each pixel does not have the same sensitivity as others to capture photons, then some will be seen in one pixel and not in the other, that is why a varying brightness level is seen. So the objective is to homogenize all the iluminated pixels. 
<h1 style="color:hotpink; font-size:13px;"> What features are recognizable and what is causing this features: </h1> In the flat field images, numerous "doughtnuts" are expected to appear due to dust, being some out of focus dust specks (such as points) present on the dewar (silvered glass with a vacuum between the walls)window and also in the filter. 

```python
lista_for_flat = []
for data_contained in listaa:
    exptime = data_contained[2]
    imagtype = data_contained[1]
    if imagtype == "Flat Field":
        lista_for_flat.append([data_contained[0],exptime,data_contained[3]])
print(lista_for_flat)
#As made before, first we need to access to the files .fit in the flat folder
folder_flat = "reduction_project/flat/"
flat_combination = []
l_flat = []
#Here two conditions will be applied; 
#First, the images have to be captured the same filter
#In second place, images with same exposure time will be considered (in this case 3 out 4 flat frames), due to if the image has more exptime can capture more dust, then it could affect the entire master flat
for file in os.listdir(folder_flat):
    with fits.open(file) as hdul:
        #Here the condition will be applied; 
        #First, the images have to be captured the same filter and also with the same exposure time, in this case I do not consider the image with exptime of 10.0 sec (this can be seen in the list below
        if (hdul[0].header["FILTER"] == "sloan_r") & (hdul[0].header["EXPTIME"] != 10.0): #bc all of them have to be with the same filter  
            flat_data = hdul[0].data
            header_flat = hdul[0].header
            plt.figure(figsize=(3,3))
            plt.imshow(flat_data, cmap="inferno",vmin =np.percentile(flat_data,5),vmax=np.percentile(flat_data,95))
            plt.title(f"Flat Field {file}")
            plt.colorbar()
            plt.show()
            #first the flats are corrected, and then normalized
            flat_calibration = (flat_data-master_bias-master_dark*(header_flat["EXPTIME"])) #when master dark is multiplicated to time, there are just ADU 
            flat_median=np.median(flat_calibration) 
            flatt = flat_calibration/flat_median
            l_flat.append(flatt)#then all flats normalized combine 
combination_mf = np.array(l_flat) #here the conversion to arrays
master_flat = np.median(combination_mf, axis=0) #as before, the median is used because data will not seen too affected because of some pixel is traslated or defectuous
fits.writeto("reduction_project/products/master_flat/master_flat.fits", master_flat, overwrite=True)
```

<h1 style="color:hotpink; font-size:18px;">⋆⭒˚.⋆ Part 4: Reduce a Science image ⋆⭒˚.⋆ </h1>

```python
#Here I look which image is light (net scientific observation)
lista_for_light= []
for data_contained in listaa: #HERE this list was made by looking into the file's header
    exptime = data_contained[2] #here the exposure time is obtained
    imagtype = data_contained[1] #here the image type is obtained
    if imagtype == "Light Frame":
        lista_for_light.append([data_contained[0],exptime])
print(lista_for_light)
```

With the list with sublists we can see that the light images are image5.fit, image9.fit and image13.fit, having the same exposition time. To get the corrected image, the image9.fit is chosen.

```python
#I selected image9.fit to reduce
comb_science = []
with fits.open("image9.fit") as hdul:
    data_light= hdul[0].data #image directly from the telescope with all 
    exptime_im = hdul[0].header["EXPTIME"]
    data_calibrate = (data_light-master_bias-master_dark*exptime_im)/master_flat #it subtracts the bias, the thermal noise, (objeto+cielo)*flat and general noise
    comb_science.append(data_calibrate)
    science_comb = np.array(comb_science)
    master_science = np.median(science_comb,axis=0)
#the .fits files are saved in their directories as made before
fits.writeto("reduction_project/raw/light_frame.fits", data_light, overwrite=True)
fits.writeto("reduction_project/products/calibrated_science/master_science.fits", master_science, overwrite=True)
```

```python
uncorrected_image = data_light #this means the image will all noise included 
corrected_image = master_science #this is the reduced image  
#here is considered the minimum and maximum contrast to the same scale in both cases, to see that the image was correctly reduced 
vmin=np.percentile(data_light,5)
vmax=np.percentile(data_light,95)
fig, axes = plt.subplots(1,2, figsize=(10,4)) #in this line the configuration to show two adjacent images was searched on the internet
#Original image
im0 = axes[0].imshow(data_light, cmap="inferno",vmin=vmin,vmax=vmax)
axes[0].set_title("Uncorrected image")
fig.colorbar(im0, ax=axes[0])
#Corrected image
im1 = axes[1].imshow(master_science, cmap="inferno",vmin=vmin,vmax=vmax)
axes[1].set_title("Corrected image")
fig.colorbar(im1, ax=axes[1])
plt.show()
```

The uncorrected image or science image (light frame) on the left includes all the data, all the visual residuals, such as vignettes, saturations, dust. While the corrected image (right) to the same scale as the uncorrected one, is noticeable that CCD corrected image frame shows a star field, obtained by using the master bias substraction, as well as the master dark (ADU/sec) multiplicated by time (sec) to simplify to ADU, and then dividing by the master flat. The background for this corrected image seems at an uniform level and distribution. It means all pixel-to-pixel nonuniformities were removed in the reduction process following the workflows above.

```python
print("Uncorrected image mean:", np.mean(data_light))
print("Uncorrected image std:", np.std(data_light))
print("Corrected image mean:", np.mean(master_science))
print("Corrected image std:", np.std(master_science))
plt.figure(figsize=(3,3.5))
plt.hist(data_light.flatten(),bins=30,color="grey",range=(80,200))
plt.title("Distribution of uncorrected image for image9.fit")
plt.xlabel("ADU")
plt.ylabel("Number of pixels")
plt.show()
plt.figure(figsize=(3,4))
plt.hist(master_science.flatten(),bins=30,color="grey",range=(-1050,-600))
plt.title("Distribution of corrected image for image9.fit")
plt.xlabel("ADU")
plt.ylabel("Number of pixels")
plt.show()
```

For the histograms of raw and science image analysis of the subtraction process are done. For the master science (corrected image) histogram we have negative values due to we substract a big negative number so the general data keeps negative too.
